### 49. Write a Python script to display the
- a. Current date and time
- b. Current year
- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
def solution():
    tod = pd.Timestamp.now()
    yr = tod.year
    mon = tod.month
    wk_num = tod.weekofyear
    wk_day = tod.day_of_week
    yr_day = tod.day_of_year
    mon_day = dt.datetime.today().timetuple().tm_mday
    wk_day1 = dt.datetime.today().strftime("%A")
    return('Current date and time', tod, 'Year',yr, 'Month',mon,'Week number of the year', wk_num, 
           'Day of week', wk_day,'Day of year', yr_day, 'Day of the month', mon_day, 'Weekday of the week', wk_day1)
     
solution()
```
